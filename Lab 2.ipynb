# Question 1: Signal Generation and Fourier Analysis

### (a) Generate a Sinusoidal Signal

This first step involves creating a simple sinusoidal signal. We define its parameters: a frequency of 5 Hz and a sampling rate of 1000 Hz. The signal is generated for a duration of 1 second.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# (a) Generate a basic sinusoidal signal
fs = 1000  # Sampling frequency in Hz
f = 5      # Frequency of the sine wave in Hz
duration = 2  # Duration of the signal in seconds
t = np.arange(0, duration, 1/fs)  # Time vector
x = np.sin(2 * np.pi * f * t)  # Sinusoidal signal

print("Signal generated successfully.")
```

### (b) Plot the Time-Domain Waveform

Here, we visualize the signal we just created. The plot shows the amplitude of the sine wave over time.

```python
# (b) Plot the time-domain waveform
plt.figure(figsize=(12, 6))
plt.plot(t, x)
plt.title('Time-Domain Waveform of the Sinusoidal Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

### (c) Compute and Plot the DTFT

The Discrete-Time Fourier Transform (DTFT) provides a continuous frequency spectrum of a discrete-time signal. Since we cannot compute the true DTFT computationally, we approximate it by computing a Discrete Fourier Transform (DFT) with a large number of points (by zero-padding the signal). This gives a high-density sampling of the DTFT, making the plot appear continuous.

```python
# (c) Compute and plot the DTFT (approximated by a zero-padded DFT)
N_dft = 8192  # Number of points for DFT, large for smooth DTFT plot
X_dtft = fft(x, N_dft)
freq_dtft = fftfreq(N_dft, 1/fs)

# Plot only the positive frequencies
positive_freq_indices = np.where(freq_dtft >= 0)
freq_dtft_pos = freq_dtft[positive_freq_indices]
X_dtft_pos = np.abs(X_dtft[positive_freq_indices])

plt.figure(figsize=(12, 6))
plt.plot(freq_dtft_pos, X_dtft_pos)
plt.title('Approximated DTFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 20)  # Zoom in to see the peak clearly
plt.show()
```

### (d) Compute and Plot the DFT

The Discrete Fourier Transform (DFT) computes the frequency spectrum of the signal at discrete frequency points. The number of points in the DFT is equal to the number of samples in the original signal. The resulting plot is a discrete spectrum, represented by stems.

```python
# (d) Compute and plot the DFT
N = len(x)  # Number of points for DFT, same as signal length
X_dft = fft(x)
freq_dft = fftfreq(N, 1/fs)

# Plot only the positive frequencies
positive_freq_indices_dft = np.where(freq_dft >= 0)
freq_dft_pos = freq_dft[positive_freq_indices_dft]
X_dft_pos = np.abs(X_dft[positive_freq_indices_dft])

plt.figure(figsize=(12, 6))
plt.stem(freq_dft_pos, X_dft_pos)
plt.title('DFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 20) # Zoom in to see the peak clearly
plt.show()
```

# Question 2: Composite Signal Analysis

### (a) Generate a Composite Signal

We will now create a composite signal by summing two sinusoidal signals with different frequencies and amplitudes.
- Signal 1: Frequency `f1 = 50 Hz`, Amplitude `A1 = 0.7`
- Signal 2: Frequency `f2 = 120 Hz`, Amplitude `A2 = 1.0`
The sampling frequency remains `fs = 1000 Hz`.

```python
# (a) Generate a composite signal
fs = 1000  # Sampling frequency
t = np.arange(0, 1, 1/fs)  # Time vector for 1 second

# Signal 1
f1 = 50    # Frequency of the first sine wave
A1 = 0.7   # Amplitude of the first sine wave
x1 = A1 * np.sin(2 * np.pi * f1 * t)

# Signal 2
f2 = 120   # Frequency of the second sine wave
A2 = 1.0   # Amplitude of the second sine wave
x2 = A2 * np.sin(2 * np.pi * f2 * t)

# Composite signal
x_composite = x1 + x2

print("Composite signal generated successfully.")
```

### (b) Plot the Time-Domain Waveform of the Composite Signal

Now, let's visualize the composite signal in the time domain. The plot will show the combined waveform of the two sinusoids.

```python
# (b) Plot the time-domain waveform of the composite signal
plt.figure(figsize=(12, 6))
plt.plot(t, x_composite)
plt.title('Time-Domain Waveform of the Composite Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

### (c) Compute and Plot the DTFT of the Composite Signal

As before, we'll approximate the DTFT by computing a zero-padded DFT. This will show the continuous frequency spectrum of our composite signal, where we expect to see peaks at 50 Hz and 120 Hz.

```python
# (c) Compute and plot the DTFT of the composite signal
N_dft_comp = 8192  # Number of points for DFT
X_dtft_comp = fft(x_composite, N_dft_comp)
freq_dtft_comp = fftfreq(N_dft_comp, 1/fs)

# Plot only the positive frequencies
positive_freq_indices_comp = np.where(freq_dtft_comp >= 0)
freq_dtft_pos_comp = freq_dtft_comp[positive_freq_indices_comp]
X_dtft_pos_comp = np.abs(X_dtft_comp[positive_freq_indices_comp])

plt.figure(figsize=(12, 6))
plt.plot(freq_dtft_pos_comp, X_dtft_pos_comp)
plt.title('Approximated DTFT Magnitude Spectrum of Composite Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 200)  # Zoom in to see the peaks clearly
plt.show()
```

### (d) Compute and Plot the DFT of the Composite Signal

Finally, we compute the DFT of the composite signal. This will give us a discrete frequency spectrum, showing the frequency components at specific, discrete points.

```python
# (d) Compute and plot the DFT of the composite signal
N_comp = len(x_composite)
X_dft_comp = fft(x_composite)
freq_dft_comp = fftfreq(N_comp, 1/fs)

# Plot only the positive frequencies
positive_freq_indices_dft_comp = np.where(freq_dft_comp >= 0)
freq_dft_pos_comp = freq_dft_comp[positive_freq_indices_dft_comp]
X_dft_pos_comp = np.abs(X_dft_comp[positive_freq_indices_dft_comp])

plt.figure(figsize=(12, 6))
plt.stem(freq_dft_pos_comp, X_dft_pos_comp)
plt.title('DFT Magnitude Spectrum of Composite Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 200) # Zoom in to see the peaks clearly
plt.show()
```

### (e) Analysis of Time-Domain vs. Frequency-Domain

The exponentially decaying signal is localized in time; it starts at a peak value and quickly diminishes. In the frequency domain, this translates to a wideband spectrum. The energy is spread across a broad range of frequencies, with the most significant component at 0 Hz (DC) and decreasing as frequency increases. This is a manifestation of the time-frequency uncertainty principle: a signal that is narrow in one domain (in this case, time) must be wide in the other (frequency). The faster the decay in the time domain (a larger 'a' value), the wider the spread of its frequency spectrum.

# Question 4: Rectangular Pulse Signal Analysis

### (a) Generate a Rectangular Pulse Signal

Next, we will generate a rectangular pulse. This signal has a constant amplitude for a finite duration and is zero otherwise. We'll define a pulse that is centered at `t=0.5s` and has a width of `0.2s`.

```python
# (a) Generate a rectangular pulse signal
fs = 1000  # Sampling frequency
t = np.arange(0, 1, 1/fs)  # Time vector for 1 second
pulse_width = 0.2
pulse_center = 0.5
x_rect = np.zeros_like(t)
x_rect[(t >= pulse_center - pulse_width/2) & (t <= pulse_center + pulse_width/2)] = 1

print("Rectangular pulse signal generated successfully.")
```

### (b) Plot the Time-Domain Waveform

Here, we visualize the rectangular pulse signal in the time domain.

```python
# (b) Plot the time-domain waveform
plt.figure(figsize=(12, 6))
plt.plot(t, x_rect)
plt.title('Time-Domain Waveform of the Rectangular Pulse')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.ylim(-0.1, 1.1)
plt.grid(True)
plt.show()
```

### (c) Compute and Plot the DTFT

The Fourier transform of a rectangular pulse in the time domain is a `sinc` function in the frequency domain. We'll approximate the DTFT to visualize this continuous spectrum.

```python
# (c) Compute and plot the DTFT
N_dft_rect = 8192
X_dtft_rect = fft(x_rect, N_dft_rect)
freq_dtft_rect = fftfreq(N_dft_rect, 1/fs)

# We need to shift the frequencies for plotting because the pulse is not centered at t=0
# This shift compensates for the time delay
shift_factor = np.exp(-1j * 2 * np.pi * freq_dtft_rect * pulse_center)
X_dtft_rect_shifted = X_dtft_rect * shift_factor


positive_freq_indices_rect = np.where(freq_dtft_rect >= 0)
freq_dtft_pos_rect = freq_dtft_rect[positive_freq_indices_rect]
X_dtft_pos_rect = np.abs(X_dtft_rect_shifted[positive_freq_indices_rect])

plt.figure(figsize=(12, 6))
plt.plot(freq_dtft_pos_rect, X_dtft_pos_rect)
plt.title('Approximated DTFT Magnitude Spectrum of Rectangular Pulse')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 50)
plt.show()
```

### (d) Compute and Plot the DFT

Now, we compute the DFT. The DFT samples the `sinc` function at discrete frequency intervals.

```python
# (d) Compute and plot the DFT
N_rect = len(x_rect)
X_dft_rect = fft(x_rect)
freq_dft_rect = fftfreq(N_rect, 1/fs)

positive_freq_indices_dft_rect = np.where(freq_dft_rect >= 0)
freq_dft_pos_rect = freq_dft_rect[positive_freq_indices_dft_rect]
X_dft_pos_rect = np.abs(X_dft_rect[positive_freq_indices_dft_rect])

plt.figure(figsize=(12, 6))
plt.stem(freq_dft_pos_rect, X_dft_pos_rect)
plt.title('DFT Magnitude Spectrum of Rectangular Pulse')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 50)
plt.show()
```

### (e) Analysis of Time-Domain vs. Frequency-Domain

The rectangular pulse is sharply defined and time-limited. Its frequency spectrum is a `sinc` function, which is infinitely wide. This again demonstrates the time-frequency uncertainty principle. A key feature of the `sinc` function is its main lobe and side lobes. The width of the main lobe is inversely proportional to the duration of the rectangular pulse. A wider pulse in the time domain results in a narrower main lobe in the frequency domain (concentrating more energy near DC), and vice-versa. This trade-off is fundamental in signal processing.
